```python
# Step 1: Import necessary Python libraries

import pandas as pd        # For loading and analyzing tabular data
import matplotlib.pyplot as plt  # For creating basic plots like bar charts and scatter plots
import seaborn as sns      # For creating advanced visualizations like heatmaps

# Optional: Set a visual style for plots to look cleaner
sns.set(style="whitegrid")

# Optional: Suppress warnings for a cleaner output view
import warnings
warnings.filterwarnings("ignore")
```

```python
# Step 2: Load the Titanic dataset

# Read the dataset using pandas
df = pd.read_csv("titanic.csv")

# Display the first 5 rows to understand the data structure
df.head()
```

```python
# Step 3: Basic exploration of the dataset

# Check the number of rows and columns
print("Dataset Shape:", df.shape)

# Get column names, data types, and non-null counts
print("\nData Information:")
df.info()

# View basic statistics for numeric columns
print("\nSummary Statistics:")
df.describe()

# üîç NEW: Check for missing values in each column
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Step 4: Calculate average of selected columns

# Calculate the average (mean) age of passengers
average_age = df['Age'].mean()
print("Average Age of Passengers:", round(average_age, 2))

# Calculate the average fare paid by passengers
average_fare = df['Fare'].mean()
print("Average Fare Paid:", round(average_fare, 2))
```

```python
# Step 5: Bar chart showing survival count by gender

plt.figure(figsize=(6,4))  # Set plot size

# Create a grouped bar chart with gender on X-axis and survival count
sns.countplot(data=df, x='Sex', hue='Survived')

# Add titles and labels
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])

# Show the plot
plt.show()
```

```python
# Step 6: Scatter plot showing Age vs Fare, with survival color-coded

plt.figure(figsize=(8,5))  # Set plot size

# Create scatter plot with survival as hue
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='Set1')

# Add title and labels
plt.title('Age vs Fare (Survival Status Highlighted)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])

# Show the plot
plt.show()
```

```python
# Step 7: Heatmap of correlation matrix for numeric columns

# Compute the correlation matrix (only for numeric features)
correlation_matrix = df.corr(numeric_only=True)

# Plot the heatmap using seaborn
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add a title
plt.title('Correlation Heatmap of Titanic Features')

# Show the plot
plt.show()
```

### üîç Final Insights & Conclusion

- The dataset contains information about Titanic passengers, including age, gender, fare, and survival status. There are some missing values in the 'Age' column, which could be handled with imputation in a more advanced analysis.
  
- The average age of passengers was approximately **29.70 years**, and the average fare paid was around **32.20 units**. This suggests a wide range of economic diversity on board.

- From the bar chart, we observed that **female passengers had a significantly higher survival rate** compared to male passengers. This aligns with the known "women and children first" evacuation policy.

- The scatter plot showed that passengers who paid **higher fares tended to survive more often**, and most survivors were clustered in the lower age range. This may indicate that younger and wealthier passengers had better access to lifeboats or cabins near safety exits.

- The correlation heatmap revealed a **moderate negative correlation between Fare and Pclass**, and a **positive correlation between Fare and Survival**, suggesting that **passenger class and ticket price were related to survival chances**.

Overall, the combination of statistical analysis and visualizations provided meaningful insights into the survival patterns of Titanic passengers.
